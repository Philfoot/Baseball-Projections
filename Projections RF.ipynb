```python
## Import packages and data ##
## focus will be on batting stats ##
import pandas as pd
import numpy as np

bat = pd.read_csv("2012-2018 Batting.csv")
print(bat.head(5))
```

```python
## What are we working with ##
list(bat)
```

```python
## Clean batting data for projections ##
## Select useful columns and convert counting stats to rate stats ##
def year_clean_bat(yr,thresh):
    temp = bat[bat.Season==yr][['Season',
        'Name',
        'Team',
        'Age',
        'G',
        'AB',
        'PA',
        'H',
        '1B',
        '2B',
        '3B',
        'HR',
        'R',
        'RBI',
        'GDP',
        'SB',
        'CS',
        'BB%',
        'K%',
        'AVG',
        'OBP',
        'SLG',
        'BABIP',
        'LD%',
        'GB%',
        'FB%',
        'IFFB%',
        'HR/FB',
        'IFH%',
        'Spd',
        'O-Swing%',
        'Z-Swing%',
        'Swing%',
        'O-Contact%',
        'Z-Contact%',
        'Contact%',
        'Zone%',
        'F-Strike%',
        'SwStr%'
        ]]
    temp['H/AB']=temp['H']/temp['AB']
    temp['1B/AB']=temp['1B']/temp['AB']
    temp['2B/AB']=temp['2B']/temp['AB']
    temp['3B/AB']=temp['3B']/temp['AB']
    temp['HR/AB']=temp['HR']/temp['AB']
    temp['R/AB']=temp['R']/temp['AB']
    temp['RBI/AB']=temp['RBI']/temp['AB']
    temp['GDP/AB']=temp['GDP']/temp['AB']
    temp['SB/AB']=temp['SB']/temp['AB']
    temp['SB%']=temp['SB']/(temp['CS']+temp['SB'])
    temp = temp.drop(columns=['H','1B','2B','3B','HR','R','RBI','GDP','SB','CS'])
    temp = temp[temp['AB']>=thresh]
    temp = temp.add_suffix(yr)
    return temp

bat18 = year_clean_bat(2018,200)
bat17 = year_clean_bat(2017,200)
bat16 = year_clean_bat(2016,200)
bat15 = year_clean_bat(2015,200)
bat14 = year_clean_bat(2014,200)
bat13 = year_clean_bat(2013,200)
bat12 = year_clean_bat(2012,200)
bat12.head()
```

```python
## Utilize prior 3 years for projections ##
## 2015 stats estimated based on 2014, 2013, 2012, etc. ##

bat15full = bat15.merge(bat14,how='left',left_on='Name2015', right_on='Name2014')
bat15full = bat15full.merge(bat13,how='left',left_on='Name2015',right_on='Name2013')
bat15full = bat15full.merge(bat12,how='left',left_on='Name2015',right_on='Name2012')
bat15full.columns = [x.strip().replace('2015', 'Estimate') for x in bat15full.columns]
bat15full.columns = [x.strip().replace('2014', 't_minus1') for x in bat15full.columns]
bat15full.columns = [x.strip().replace('2013', 't_minus2') for x in bat15full.columns]
bat15full.columns = [x.strip().replace('2012', 't_minus3') for x in bat15full.columns]

bat16full = bat16.merge(bat15,how='left',left_on='Name2016', right_on='Name2015')
bat16full = bat16full.merge(bat14,how='left',left_on='Name2016',right_on='Name2014')
bat16full = bat16full.merge(bat13,how='left',left_on='Name2016',right_on='Name2013')
bat16full.columns = [x.strip().replace('2016', 'Estimate') for x in bat16full.columns]
bat16full.columns = [x.strip().replace('2015', 't_minus1') for x in bat16full.columns]
bat16full.columns = [x.strip().replace('2014', 't_minus2') for x in bat16full.columns]
bat16full.columns = [x.strip().replace('2013', 't_minus3') for x in bat16full.columns]

bat17full = bat17.merge(bat16,how='left',left_on='Name2017', right_on='Name2016')
bat17full = bat17full.merge(bat15,how='left',left_on='Name2017',right_on='Name2015')
bat17full = bat17full.merge(bat14,how='left',left_on='Name2017',right_on='Name2014')
bat17full.columns = [x.strip().replace('2017', 'Estimate') for x in bat17full.columns]
bat17full.columns = [x.strip().replace('2016', 't_minus1') for x in bat17full.columns]
bat17full.columns = [x.strip().replace('2015', 't_minus2') for x in bat17full.columns]
bat17full.columns = [x.strip().replace('2014', 't_minus3') for x in bat17full.columns]

bat18full = bat18.merge(bat17,how='left',left_on='Name2018', right_on='Name2017')
bat18full = bat18full.merge(bat16,how='left',left_on='Name2018',right_on='Name2016')
bat18full = bat18full.merge(bat15,how='left',left_on='Name2018',right_on='Name2015')
bat18full.columns = [x.strip().replace('2018', 'Estimate') for x in bat18full.columns]
bat18full.columns = [x.strip().replace('2017', 't_minus1') for x in bat18full.columns]
bat18full.columns = [x.strip().replace('2016', 't_minus2') for x in bat18full.columns]
bat18full.columns = [x.strip().replace('2015', 't_minus3') for x in bat18full.columns]

batfull = bat15full.append(bat16full)
batfull = batfull.append(bat17full)
batfull = batfull.append(bat18full)
```

```python
## Final dataset only includes players with three consecutive years of 200+ ABs
batfullclean = batfull.dropna(axis=0,how='any')
```

```python
## convert data to arrays for skicit
# We will project batting average 
dep = np.array(batfullclean['AVGEstimate'])
# Restrict features to exclude the prediction year
estimate = batfullclean.columns.str.contains('.*Estimate')
features_df = batfullclean.loc[:,~estimate]
## drop a couple irrelevant/string variables
features_df = features_df.drop(columns=['Seasont_minus1', 'Seasont_minus2', 'Seasont_minus3'])
features_df = features_df.drop(columns=['Teamt_minus1', 'Teamt_minus2', 'Teamt_minus3'])
features_df = features_df.drop(columns=['Namet_minus1', 'Namet_minus2', 'Namet_minus3'])
feature_list = list(features_df.columns)
features = np.array(features_df)
```

```python
# Create training and test sets
from sklearn.model_selection import train_test_split
# Split the data into training and testing sets
train_features, test_features, train_dep, test_dep = train_test_split(features, dep, test_size = 0.25, random_state = 23)
```

```python
# Import the model we are using
from sklearn.ensemble import RandomForestRegressor
# Instantiate model with 1000 decision trees
rf = RandomForestRegressor(n_estimators = 1000, random_state = 23)
# Train the model on training data
rf.fit(train_features, train_dep)
```

```python
# Use the forest's predict method on the test data
predictions = rf.predict(test_features)
# Calculate the absolute errors
errors = abs(predictions - test_dep)
# Print out the mean absolute error (mae)
print('Mean Absolute Error:', round(np.mean(errors), 5), 'points of batting average.')
```

```python
## Compare with simple 3 year average estimate
index1 = feature_list.index("AVGt_minus1")
index2 = feature_list.index("AVGt_minus2")
index3 = feature_list.index("AVGt_minus3")

control = (test_features[:,index1]+test_features[:,index2]+test_features[:,index3])/3
baseline = abs(control - test_dep)
print('Baseline Error:', round(np.mean(baseline),5),'points of batting average.')
```

```python
## Try a baseline that weights the prior year most

control = (test_features[:,index1]*.5+test_features[:,index2]*.3+test_features[:,index3]*.2)
baseline = abs(control - test_dep)
print('Baseline Error:', round(np.mean(baseline),5),'points of batting average.')
```
